```python
def perceptron_neuron(is_sunny, is_tired):
    x0, x1, x2 = 1, is_sunny, is_tired

    w1 = 2.0 # Importance of it being sunny
    w2 = -1.5 # We don't want to be tired

    w0 = -1.0 # Theta is 1.0

    # Caluclate z, the net sum function
    z = (w0 * x0) + (w1 * x1) + (w2 * x2)
    output = 1 if z >= 0 else -1

    return output, z

# Testing
decision, net_input = perceptron_neuron(1, 1)
print("Net Input (z): ", net_input)
print("Decision: ", decision)


```
